```python
import numpy as np
import pandas as pd
import sklearn
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection and Pre processing


```python
#loading the data from a csv file to a pandas dataframe
movies_data = pd.read_csv('/content/movies.csv')
```

```python
#printing the first five rows of the dataframe
movies_data.head()
```

```python
#finding the number of rows and coloums in the data frame
movies_data.shape
```

```python
#selecting the relevant features for recommendation
selected_features= ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
#replacing the null values with null string
for feature in selected_features:
 movies_data [feature] = movies_data[feature].fillna('')
```

```python
#combination of all the selected features
combined_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_features)
```

```python
#now we have to convert the tyext datas to feature vectors
vectorizer=TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

cosine answer


```python
#getting the similarity scores using cosine similaritity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
#getting the movie name from the user
movie_name = input('enter your fav movie name:')
```

```python
#creating a list with all movie names given in the dataset
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the best match for the movie name given by the user
find_best_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_best_match)
```

```python
best_match = find_best_match[0]
print(best_match)
```

```python
#find the index of the movie with title
index_of_the_movie = movies_data[movies_data.title == best_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting a list of similar movies
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#swaping the movies on their scores of similarity
sorted_similar_movies = sorted(similarity_score,key = lambda x:x[1],reverse = True)
print(sorted_similar_movies)
```

```python
#print the name of the similar movies based on the index
print('movies suggested for the viewers:\n')
i = 1
for movies in sorted_similar_movies:
  index = movies[0]
  title_from_index = movies_data[movies_data.index ==index]['title'].values[0]
  if(i<30):
    print(i,'.',title_from_index)
  i+=1
```

Movies prediction system
